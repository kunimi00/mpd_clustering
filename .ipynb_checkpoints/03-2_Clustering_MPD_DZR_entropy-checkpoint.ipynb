```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pickle
from sklearn.cluster import KMeans, SpectralClustering
from sklearn.manifold import TSNE
from sklearn import preprocessing
from util.util_data import *

np.set_printoptions(suppress=True)
```

```python
def save_obj_curr_folder(obj, name):
    with open(name, 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj_curr_folder(name):
    with open(name, 'rb') as f:
        return pickle.load(f)
```

```python
pl_X_dict = load_json_to_dict('data_pl_dzrmeta/pl_X_dict.json')
pl_Y_title_dict = load_json_to_dict('data_pl_dzrmeta/pl_Y_title_dict.json')

pl_X_list = load_txt_to_list('data_pl_dzrmeta/pl_X_list.txt')
pl_key_to_id_dict = load_json_to_dict('data_pl_dzrmeta/pl_key_to_id_dict.json')

# pl_X = load_obj_curr_folder('data_pl_dzrmeta/pl_X.p')
# pl_Y_desc = load_obj_curr_folder('data_pl_dzrmeta/pl_Y_desc.p')
# pl_Y_title = load_obj_curr_folder('data_pl_dzrmeta/pl_Y_title.p')
```

```python
pl_X = []
for _item in pl_X_list:
    pl_X.append(eval(_item))
```

```python
pl_X[0]
```

```python
pl_X = np.array(pl_X)
```

```python
pl_X.shape
```

```python
std_pl_X = preprocessing.scale(pl_X, axis=0) 
```

```python
A = std_pl_X
interia_ls = []
for k in range (1, 20):
    # Create a kmeans model on our data, using k clusters.  random_state helps ensure that the algorithm returns the same results each time.
    kmeans_model = KMeans(n_clusters=k, random_state=1).fit(A)

    # These are our fitted labels for clusters -- the first cluster has label 0, and the second has label 1.
    labels = kmeans_model.labels_

    # Sum of distances of samples to their closest cluster center
    interia = kmeans_model.inertia_
    interia_ls.append(interia)
    print("k:",k, " distance to closest centroid:", interia)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(range(1, 20), interia_ls);
plt.plot(range(1, 20), interia_ls);
```

```python
# try different initialization states to verify the clustering results

inertia_ls = []
centers_ls = []
labels_ls = []
for _rs in range (1, 10):
    kmeans_model = KMeans(n_clusters=8, random_state=_rs).fit(A)

    labels = kmeans_model.labels_
    inertia = kmeans_model.inertia_
    centers = kmeans_model.cluster_centers_

    labels_ls.append(labels)
    inertia_ls.append(inertia)
    centers_ls.append(centers)
    
    print("_rs:",_rs, " distance to closest centroid:", inertia)
```

```python
centers_ls = np.array(centers_ls)
```

```python
centers_ls.shape
```

```python
centers_ls_all = []
for _i in range(centers_ls.shape[0]):
    centers_ls_all.extend(centers_ls[_i])
centers_ls_all = np.array(centers_ls_all)
```

```python
centers_ls_all.shape
```

```python
tsne_different_centers = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_pl_X_different_centers = tsne_different_centers.fit_transform(centers_ls_all)
```

```python
y_rs = []
for _i in range(9):
    y_rs.extend([_i for _ in range(8)])
y_rs = np.array(y_rs)
```

```python
print(y_rs)
```

```python
plt.figure(figsize=(16,10))
plt.scatter(tsne_pl_X_different_centers[:, 0], tsne_pl_X_different_centers[:, 1], c=y_rs, s=100, cmap='plasma')
```

```python
feature_name_list = ['category entropy',
                     'decade entropy',
                     'genre entropy',
                     'instrument entropy',
                     'lang entropy',
                     'location entropy',
                     'location:continent entropy',
                     'mood entropy',
                     'mood:activity entropy',
                     'mood:emotion entropy',
                     'mood:situation entropy',
                     'singer_type entropy',
                     'num_albums',
                     'num_artists',
                     'num_tracks',
                     'num_followers'
                     ]
```

```python
for c_idx in range(len(centers)):
    
    x = centers[c_idx]
    n_groups = len(feature_name_list)

    
    fig, ax = plt.subplots(figsize=(20, 5))
    for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] + ax.get_xticklabels() + ax.get_yticklabels()):
        item.set_fontsize(16)

    index = np.arange(n_groups)
    bar_width = 0.30

    opacity = 0.4
    error_config = {'ecolor': '0.3'}

    
    rects1 = ax.bar(index, x, bar_width,
                    alpha=opacity, color='rgbkymc', 
                    error_kw=error_config,
                    label='playlist features')
    ax.axhline(0, color='gray')
    
    
    ax.set_ylim(bottom=-5, top=5)
    ax.set_xlabel(str(c_idx))
    ax.set_ylabel(str(c_idx))
    ax.set_xticks(index + bar_width / 2)
    ax.set_xticklabels(feature_name_list, rotation=60)
    ax.legend()
    fig.tight_layout()
    plt.savefig('./data_pl_dzrmeta/' + str(c_idx) + '.png')
```
